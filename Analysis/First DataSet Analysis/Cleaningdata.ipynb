```python
# renaming the columns name
data_rename={
'AGERNG':'age',
'GENDER':'gender',
'EDU':'education',
'PROF':'profession',
'MARSTS':'maritial status',
'RESDPL':'residential place',
'LIVWTH':'living with',
'DEBT':'debt',
'SMOKE':'smoke',
'DRINK':'drink',
'ILLNESS':'illness',
'AVGSLP':'average sleep',
'INSOM':'insomia',
'WRKPRE':'work pressure',
'ANXI':'anxiety',
'DEPRI':'depressed',
'ABUSED':'abused',
'CHEAT':'cheat',
'THREAT':'threat',
'SUICIDE':'sucide',
'LOST':'lost',
'DEPRESSED':'depression',
}
data.rename(columns=data_rename,inplace=True)
data.head()
```

```python
# Now we will convert categorical data in to numeric form
age_dict={
'26-30':1,
'21-25':2,
'16-20':3,
'31-35':4,
'46-50':5,
'41-45':6,
'56-60':7,
'36-40':8,
'61+':9,
'51-55':10,
}
gender_dict={
    'Female':1,
    'Male':2
}
education_dict={
'Post Graduate':1,
'HSC':2,
'Graduate':3,
'SSC':4,
}
profession_dict={
'Unemployed':1,
'Service holder (Private)':2,
'Student':3,
'Service holder (Government)':4,
'Other':5,
'Businessman':6,
}
maritial_status_dict={
'Unmarried':1,
'Married':2,
'Divorced':3,
}
residential_place_dict={
'Town':1,
'City':2,
'Village':3,
}
living_with_dict={
'With Family':1,
'Without Family':2,
}
debt_dict={
'Yes':1,
'No':2
}
smoke_dict={
'Yes':1,
'No':2
}
drink_dict={
'Yes':1,
'No':2
}
illness_dict={
'Yes':1,
'No':2
}
average_sleep_dict={
'More than 8 hours':1,
'6 hours':2,
'8 hours':3,
'7 hours':4,
'5 hours':5,
'Below 5 hours':6,
}
insomia_dict={
'Yes':1,
'No':2
}
work_pressure_dict={
'No Pressure':1,
'Moderate':2,
'Mild':3,
'Severe':4,
}
anxiety_dict={
'Yes':1,
'No':2,
}
depressed_dict={
'Yes':1,
'No':2,
}
abused_dict={
'Yes':1,
'No':2,
}
cheat_dict={
'Yes':1,
'No':2,
}
threat_dict={
'Yes':1,
'No':2,
}
sucide_dict={
'Yes':1,
'No':2,
}
lost_dict={
'Yes':1,
'No':2,
}
```

```python
len(data.columns)
```

```python
data['age']=data['age'].map(age_dict)
data['gender']=data['gender'].map(gender_dict)
data['education']=data['education'].map(education_dict)
data['profession']=data['profession'].map(profession_dict)
data['maritial status']=data['maritial status'].map(maritial_status_dict)
data['residential place']=data['residential place'].map(residential_place_dict)
data['living with']=data['living with'].map(living_with_dict)
data['debt']=data['debt'].map(debt_dict)
data['smoke']=data['smoke'].map(smoke_dict)
data['drink']=data['drink'].map(drink_dict)
data['illness']=data['illness'].map(illness_dict)
data['average sleep']=data['average sleep'].map(average_sleep_dict)
data['insomia']=data['insomia'].map(insomia_dict)
data['work pressure']=data['work pressure'].map(work_pressure_dict)
data['anxiety']=data['anxiety'].map(anxiety_dict)
data['depressed']=data['depressed'].map(depressed_dict)
data['abused']=data['abused'].map(abused_dict)
data['cheat']=data['cheat'].map(cheat_dict)
data['threat']=data['threat'].map(threat_dict)
data['sucide']=data['sucide'].map(sucide_dict)
data['lost']=data['lost'].map(lost_dict)
```
