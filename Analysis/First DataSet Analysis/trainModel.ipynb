# Seperating target column and remaining column for constructing model

```python
target=data['depression']
inputs=data.drop("depression",axis=1)
len(inputs.columns)
```

# importing modules to spilt the data

```python
from sklearn.model_selection import train_test_split
np.random.seed(10)
X_test,X_train,y_test,y_train=train_test_split(inputs,target,test_size=.25)
```

# Importing all the modules required for classification and for model Evaluation

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

# Making dictionary for each model so that we don't have to do it manually

```python
models = { "Decision Tree Classifier": DecisionTreeClassifier(random_state=99),
          "KNN": KNeighborsClassifier(),
          "Random Forest Classification": RandomForestClassifier(random_state=42),
          "Logistic Regression": LogisticRegression(),
          "Support Vector machine":SVC(),
          "XGBoost Classifier ":XGBClassifier(use_label_encoder=False)}
```

# Making a function to return the accuracy of all the models which we have mentioned in our dictionary(models)

```python
def find_accuracy(X_test,X_train,y_test,y_train):
    model_scores={}
    for name,model in models.items():
        np.random.seed(10)
        model.fit(X_train,y_train)
        model_scores[name]=model.score(X_test,y_test)
    return model_scores
    
    
    
```

```python
accuracy=find_accuracy(X_test,X_train,y_test,y_train)
accuracy
```

# We can see that Random forest classification has the maxiumn accuracy Hence we are going ahead with this

```python
model=RandomForestClassifier(random_state=42)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
```

```python
accuracy = (cm[0][0]+cm[1][1])/(cm[0][1] + cm[1][0] +cm[0][0] +cm[1][1])
print(accuracy*100)
```
